Seaborn

* Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_theme()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', col= 'time', hue= 'smoker', size= 'size')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

Scatter PLot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y= 'petal_width', hue='species', data=iris)
```

Loading titanic dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

Bar chart

```python
sns.barplot(x='sex', y='survived',hue= 'class', data=titanic)
```

```python
sns.distplot(titanic['fare'])
```

Correlation
* Positive Correlation
* Negatve Correlation

```python
df_titanic = titanic.drop(['sex', 'embarked', 'class', 'who','deck', 'embark_town'], axis=1)
df_titanic.replace({'alive':{'yes':1, 'no':0}}, inplace=True)
df_titanic.head()
correaltion = df_titanic.corr()
print(correaltion)
plt.figure(figsize=(10,10))
sns.heatmap(correaltion, cbar=True, square=True, fmt='.1f', annot=True, cmap='Blues')
```
