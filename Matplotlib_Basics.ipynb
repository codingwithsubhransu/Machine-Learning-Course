Matplotlib


*   Useful for making plots


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the above data

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

Adding title x-axis and y-axis

```python
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine wave')
plt.title('sine wave')
plt.show()
```

```python
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ["English", "hindi", "French", "German"]
students = [23, 17, 35, 29]
ax.bar(languages, students)
plt.xlabel('languages')
plt.ylabel('Number of students')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ["English", "hindi", "French", "German"]
students = [23, 17, 35, 29]
ax.pie(students, labels = languages, autopct= '1%.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color = 'g')
ax.scatter(x, z, color = 'r')
plt.show()
```

3d Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap = 'Greens')
plt.show()
```
