#### Imbalanced dataset
A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/drive/MyDrive/Datasets/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
credit_card_data.tail()
```

```python
credit_card_data['Class'].value_counts()
```

0 -> Legit transactions

1 -> Fradulant data

# This is highly imbalanced dataset

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Building a sample dataset containing similar distribution of Legit and Fradulant Transactions

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

Concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.tail()
```
